```python
data = []
last = -1, -1

with open('./wisdm.txt/WISDM_ar_v1.1_raw.txt') as inp:
        lines = inp.read().split(';')
        
        for line in lines:
            tokens = line.strip().split(',')
            tokens = [t for t in tokens if t.strip()!='']
            if len(tokens)!=6:
                print 'bad line', line
                continue
            if (tokens[0], tokens[1]) == last:
                data[-1].append([float(tokens[-3]), float(tokens[-2]), float(tokens[-1])])
                
            else:
                last = tokens[0], tokens[1]
                data.append([])
                data[-1].append([float(tokens[-3]), float(tokens[-2]), float(tokens[-1])])
```

```python
for d in data:
    print len(d)
```

```python
X, Y = [], []
for d in data:
    
    for i in xrange(0, len(d)-200, 200):
        X.append(d[i:i+199])
        Y.append(d[i+199])
   
```

```python
X, Y = [], []
for d in data:
    
    for i in xrange(0,len(d)-1,2):
            
            
            X.append(d[i])
            Y.append(d[i+1])
```

```python
import numpy as np
X = np.array(X)
Y = np.array(Y)
print X.shape, Y.shape
```

```python
X = np.array(X)
X = np.array([x.flatten() for x in X])#
#X = np.linalg.norm(X,  axis=2)
mean_ = np.mean(np.mean(X))
std_ = np.std(X)
X = ( X- mean_) / std_

Y = np.array(Y)
Y = ( Y- mean_) / std_

Y = np.linalg.norm(Y,  axis=1)
#since these all are independent time chunks
print X.mean(), X.std(), Y.mean(), Y.std()


```
